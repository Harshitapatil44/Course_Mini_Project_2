## Task 1: Data Preparation
Loading and transforming the transactional data for the retail context.

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
import time

# Load CSV
df = pd.read_csv('cleaned_retail_sales_data.csv')

# Convert to SQLite
conn = sqlite3.connect(':memory:')  # Using in-memory for the demo
df.to_sql('sales_data', conn, index=False, if_exists='replace')
print('Data loaded into SQLite successfully.')
```

## Task 2: Chatbot Core Analytics Engine
Backend logic to interpret queries and provide insights.

```python
def run_query(sql_query):
    return pd.read_sql_query(sql_query, conn)

def analyze_query(user_prompt):
    # Simulating intent classification (Descriptive vs Visualization)
    if 'plot' in user_prompt.lower() or 'chart' in user_prompt.lower():
        return 'Visualization'
    return 'Descriptive'

print('Analytics engine initialized.')
```

## Task 3: GenAI Integration (Simulated)
Integrating LLM to parse natural language to SQL.

```python
def get_ai_response(prompt):
    # In a real app, this calls GPT-4 API
    start_time = time.time()
    if 'top categories' in prompt.lower():
        query = 'SELECT ProductCategory, SUM(TransactionAmount) as TotalSales FROM sales_data GROUP BY ProductCategory ORDER BY TotalSales DESC LIMIT 5'
        data = run_query(query)
        response = f'The top 5 product categories are {list(data['ProductCategory'])}.'
    else:
        response = 'I can help you with sales trends, categories, and performance. Try asking: Top categories by sales?'
    
    processing_time = round(time.time() - start_time, 2)
    return response, processing_time

res, pt = get_ai_response('What are the top categories?')
print(f'AI Response: {res} (Time: {pt}s)')
```

## Task 4: UI Development (Interface Mockup)
Using ipywidgets to demonstrate the chat interface.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

chat_history = []
output = widgets.Output()

def on_send_clicked(b):
    prompt = user_input.value
    response, pt = get_ai_response(prompt)
    chat_history.append({'prompt': prompt, 'response': response, 'time': pt})
    with output:
        clear_output()
        for chat in chat_history:
            print(f'User: {chat['prompt']}')
            print(f'AI: {chat['response']} [{chat['time']}s]')
            print('-'*30)

user_input = widgets.Text(placeholder='Type your sales question here...')
send_button = widgets.Button(description='Send')
send_button.on_click(on_send_clicked)

display(user_input, send_button, output)
```
